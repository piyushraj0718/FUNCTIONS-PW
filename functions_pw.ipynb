# QUESTION 1  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_of_evens(numbers):
    return sum(num for num in numbers if num % 2 == 0)

nums = [11,12, 13, 14, 15, 16]
result = sum_of_evens(nums)
print(result)
```

# QUESTION 4 Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

def prime_numbers_up_to_200():
    primes = [num for num in range(1, 201) if prime(num)]
    return primes

# Example usage
print(prime_numbers_up_to_200())
```

# QUESTION 5  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of TERMS.

```python
def fibonacci(n_terms):
    a, b = 0, 1
    for _ in range(n_terms):
        yield a
        a, b = b, a + b

for num in fibonacci(10):
    print(num, end=" ")
```

# QUESTION 6 Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
class PowerOfTwoIterator:
    def __init__(self, max_exponent):
        self.max_exponent = max_exponent
        self.current_exponent = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_exponent > self.max_exponent:
            raise StopIteration

        result = 2 ** self.current_exponent
        self.current_exponent += 1
        return result


power_iter = PowerOfTwoIterator(5)
for num in power_iter:
    print(num, end=" ")
```

# QUESTION 7 Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")


file_path = "C:\\Users\\YourName\\Documents\\example.txt"
for line in read_file_line_by_line(file_path):
    print(line)
```

# QUESTION 8 . Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# QUESTION 9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python

celsius_temps = [0, 20, 30, 40, 100]

fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))


print(f"Temperatures in Fahrenheit: {fahrenheit_temps}")
```

# QUESTION 10 Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python

is_not_vowel = lambda ch: ch.lower() not in 'aeiou'


input_string = "Hello, my name is piyush raj and i am from iit patna"

filtered_string = "".join(filter(is_not_vowel, input_string))

print("String after removing vowels:", filtered_string)
```
